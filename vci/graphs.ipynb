```python
import numpy as np
from matplotlib import pyplot as plt
import json
```

```python
def read_json(file_path):
	file = open(file_path)
	json_file = json.load(file)
	file.close()
	return json_file
```

```python
filepath1 = "../output/vci/03_08_01_no_mask/per_view.json"
filepath2 = "../output/vci/03_08_01_2x_fgr/per_view.json"
filepath3 = "../output/vci/03_08_01_3x_fgr/per_view.json"

data1 = read_json(filepath1)
data2 = read_json(filepath2)
data3 = read_json(filepath3)
```

```python
def get_sorted_list(data):
	a = sorted(data.items(), key=lambda x: x[0])
	return [e[1] for e in a]
```

```python
psnr1 = get_sorted_list(data1["ours_60000"]["PSNR"])
psnr2 = get_sorted_list(data2["ours_60000"]["PSNR"])
psnr3 = get_sorted_list(data3["ours_60000"]["PSNR"])
```

```python
plt.figure(figsize=(14, 4))
plt.plot(psnr1, label="alpha = 1")
plt.plot(psnr2, label="alpha = 2")
plt.plot(psnr3, label="alpha = 3")
plt.legend()
plt.grid()
plt.xlabel("Frame")
plt.ylabel("PSNR")
plt.xlim((0, 325))
plt.show()
```

```python
filepath0 =  "../output/vci/03_09_01_batch_size/per_view.json"
data0 = read_json(filepath0)
```

```python
psnr0 = get_sorted_list(data0["ours_60000"]["PSNR"])
```

```python
plt.figure(figsize=(14,4))
plt.plot(psnr1, label="Batch size = 1")
plt.plot(psnr0, label="Batch size = 2")
plt.plot(psnr3, label="Alpha = 3", linestyle=":")
plt.grid()
plt.xlim((0, 325))
plt.xlabel("Frame")
plt.ylabel("PSNR")
plt.legend()
plt.show()
```
